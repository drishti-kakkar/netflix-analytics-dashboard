# Netflix Analytics Project

This project analyzes Netflix content trends, ratings, countries, genres, and release patterns using Python, Pandas, and Plotly.

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
netflix = pd.read_csv("../dataset/netflix_titles.csv")

netflix
```

```python
netflix.columns
```

```python
netflix.isnull().sum()
```

```python
netflix["type"].value_counts()
```

```python
fig = px.pie(
    netflix,
    names="type",
    title="Movies vs TV Shows on Netflix"
)

fig.show(renderer="browser")
```

# Data Cleaning

```python
netflix.isnull().sum()
```

```python
netflix = netflix.drop_duplicates()

netflix.shape
```

```python
shape
```

```python
netflix["country"] = netflix["country"].fillna("Unknown")
```

```python
netflix["director"] = netflix["director"].fillna("Not Available")
```

```python
netflix.isnull().sum()
```

# Netflix Content Analysis

```python
netflix["rating"].value_counts()
```

```python
rating_counts = netflix["rating"].value_counts().head(10)

fig = px.bar(
    x=rating_counts.index,
    y=rating_counts.values,
    title="Top Netflix Ratings"
)

fig.show(renderer="browser")
```

```python
year_data = netflix["release_year"].value_counts().sort_index()

year_data
```

```python
fig = px.line(
    x=year_data.index,
    y=year_data.values,
    title="Netflix Content Growth Over Years"
)

fig.show(renderer="browser")
```

```python
top_countries = netflix["country"].value_counts().head(10)

top_countries
```

```python
fig = px.bar(
    x=top_countries.index,
    y=top_countries.values,
    title="Top Countries Producing Netflix Content"
)

fig.show(renderer="browser")
```

```python
netflix["listed_in"].value_counts().head(10)
```

```python
top_genres = netflix["listed_in"].value_counts().head(10)

fig = px.bar(
    x=top_genres.index,
    y=top_genres.values,
    title="Top Netflix Genres"
)

fig.show(renderer="browser")
```

```python
movies = netflix[netflix["type"] == "Movie"]

movies
```

```python
movies.shape
```

```python
movies["duration"].head()
```

```python
movies["duration"] = movies["duration"].str.replace(" min", "")
```

```python
movies["duration"] = movies["duration"].astype(float)
```

```python
movies["duration"].head()
```

```python
movies.nlargest(10, "duration")[
    ["title", "duration"]
]
```

```python
fig = px.histogram(
    movies,
    x="duration",
    nbins=30,
    title="Netflix Movie Duration Distribution"
)

fig.show(renderer="browser")
```

```python
movies["duration"].mean()
```

# Key Performance Indicators (KPIs)


```python
total_titles = netflix.shape[0]

total_movies = netflix[netflix["type"] == "Movie"].shape[0]

total_tvshows = netflix[netflix["type"] == "TV Show"].shape[0]

average_movie_duration = movies["duration"].mean()

print("Total Titles:", total_titles)

print("Total Movies:", total_movies)

print("Total TV Shows:", total_tvshows)

print("Average Movie Duration:", average_movie_duration)
```

```python
netflix["director"].value_counts().head(10)
```

```python
top_directors = netflix["director"].value_counts().head(10)

fig = px.bar(
    x=top_directors.index,
    y=top_directors.values,
    title="Top Netflix Directors"
)

fig.show(renderer="browser")
```
